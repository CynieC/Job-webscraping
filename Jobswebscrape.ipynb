```python
#import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.flexjobs.com/search?search=data+analyst%2C+machine+learning%2Cpython'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
job_listings = soup.find_all('li', attrs = {'class' : 'm-0 row job'})
len(job_listings)
```

```python
all_jobs=[]
for job in job_listings:

    job_title = job.find('a', attrs = {'class' : 'job-title job-link'}).text.strip()#get jobtitle
    job_location =job.find('div', attrs = {'class' : 'job-locations'}).text.strip() #get location of job
    job_type =job.find('div', attrs = {'class' : 'job-tags'}).text.strip() #get job type
    job_description = job.find('div', attrs = {'class' : 'job-description'}).text.strip() #gt job description

    all_jobs.append({
        'title' : job_title,
        'location' : job_location,
        'type' : job_type,
        'description' : job_description
    })
  
df = pd.DataFrame(all_jobs)
df
```

```python
#store in a dataframe
df = pd.DataFrame(all_jobs)
df.columns = ["title", "location", "type","description"]
df.head()
```
